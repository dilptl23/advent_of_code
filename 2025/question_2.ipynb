### Question 2

https://adventofcode.com/2025/day/2

```python
def is_invalid_password(password: str) -> bool:
    str_len = len(str(password))
    if str_len % 2 != 0:
        return False
    
    half_str = str(password)[:int(str_len/2)]
    if half_str == str(password)[int(str_len/2):]:
        return True
    
```

```python
with open('question_2_input.txt', 'r') as f:
    for line in f:
        values = line.strip().split(',')

all_invalid_passwords = []
for value_range_str in values:
    value_range = [int(i) for i in value_range_str.split("-")]
    for i in range(value_range[0], value_range[1] + 1):
        if is_invalid_password(i):
            all_invalid_passwords.append(i)

print("Sum of all invalid passwords:", sum(all_invalid_passwords))
```

### Question 2 Part 2

```python
# need a function that returns true if number is made up of repeating numbers
# ex: 1111, 1212, 123123
import re

password = "9995"

def has_repeating_pattern(password: str):
    end_idx = 0
    while end_idx < len(password) // 2:
        if len(password) % (end_idx + 1) != 0:
            end_idx += 1    
            continue
        matches_start_indices = [match.start() for match in re.finditer(password[:end_idx+1], password)]
        num_of_possible_matches = len(password) // (end_idx + 1)
        if len(matches_start_indices) == num_of_possible_matches:
            return True
        end_idx += 1
    return False

# LLM Solution
def is_repeating(n: int) -> bool:
    s = str(n)
    return any(s == s[:i] * (len(s) // i) 
               for i in range(1, len(s)) 
               if len(s) % i == 0)
```

```python
with open('question_2_input.txt', 'r') as f:
    for line in f:
        values = line.strip().split(',')

all_invalid_passwords = []
for value_range_str in values:
    value_range = [int(i) for i in value_range_str.split("-")]
    for i in range(value_range[0], value_range[1] + 1):
        if has_repeating_pattern(str(i)):
            all_invalid_passwords.append(i)

print("Sum of all invalid passwords:", sum(all_invalid_passwords))
```
