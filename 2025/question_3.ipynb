#### Notes

Find the highest digit not counting the last digit 
then find the highest remaining digit after that digit

```python
num = "234234234234278"


def get_joltage(num: str):
    tmp = num[:len(num)-1]
    first_digit = max([int(i) for i in tmp])
    first_digit_pos = num.find(str(first_digit))
    tmp = num[first_digit_pos+1:]
    second_digit = max([int(i) for i in tmp])
    joltage = int(str(first_digit) + str(second_digit))
    return joltage
```

```python
with open('question_3_input.txt', 'r') as f:
    all_joltages = []
    for line in f:
        all_joltages.append(get_joltage(line.strip()))

print(sum(all_joltages))
```

#### Part 2

Largest 12 digit number
- Pick the largest digit of the first len(num) - 11 digits
- Pick the largest digit of the remaining len(num) - 10 digits
- repeat

```python
num = "818181911112111"
largest_n = 12

def get_joltage(num: str, largest_n: int):
    digits = []
    for remaining_needed_digits in range(largest_n-1, -1, -1):
        tmp = num[:len(num)-remaining_needed_digits]
        digits.append(max(tmp))
        num = num[num.find(digits[-1])+1:]
    joltage = ""
    for digit in digits:
        joltage += digit    
    return int(joltage)
```

```python
with open('question_3_input.txt', 'r') as f:
    all_joltages = []
    for line in f:
        all_joltages.append(get_joltage(line.strip(), largest_n=12))

print(sum(all_joltages))
```
