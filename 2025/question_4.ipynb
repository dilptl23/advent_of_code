### Question 4

https://adventofcode.com/2025/day/4

#### Notes

- Pull input into an array of arrays where each array is a row
- Create a function that takes in the full matrix and a coordinate. This function should return True or False whether the forklift can reach this roll, meaning the function checks the 8 adjacent positions to count how many roles there are

```python
with open('question_4_input.txt', 'r') as f:
    grid = []
    for line in f:
        row = [i for i in line.strip()]
        grid.append(row)
```

```python
## function to determine if position is accessible by the fork lift
pos = (0, 7)

# function to get an array of all adjacent positions
def get_adjecent_positions(pos):
    adjacent_positions = [
        (pos[0]-1, pos[1]-1),
        (pos[0]-1, pos[1]),
        (pos[0]-1, pos[1]+1),
        (pos[0], pos[1]-1),
        (pos[0], pos[1]+1),
        (pos[0]+1, pos[1]-1),
        (pos[0]+1, pos[1]),
        (pos[0]+1, pos[1]+1)
    ]

    return adjacent_positions

def is_position_accessible(pos, grid):
    if grid[pos[0]][pos[1]] != '@':
        return False
    adj_positions = get_adjecent_positions(pos)
    total_adj_rolls = 0
    for adj_pos in adj_positions:
        if adj_pos[0] < 0 or adj_pos[0] >= len(grid) or adj_pos[1] < 0 or adj_pos[1] >= len(grid[0]):
            continue
        if grid[adj_pos[0]][adj_pos[1]] == '@':
            total_adj_rolls += 1
    if total_adj_rolls < 4:
        return True
    return False        
```

```python
total_accessible_positions = 0
for i in range(len(grid)):
    for j in range(len(grid[0])):
        if is_position_accessible((i, j), grid):
            total_accessible_positions += 1
print(total_accessible_positions)
```

### Part 2

```python
continue_loop = True
total_accessible_positions = 0
while continue_loop:
    accessible_positions_this_loop = 0
    accessible_positions = []
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            if is_position_accessible((i, j), grid):
                accessible_positions.append((i, j))
                accessible_positions_this_loop += 1

    for pos in accessible_positions:
        grid[pos[0]][pos[1]] = '.'

    if accessible_positions_this_loop == 0:
        continue_loop = False
    total_accessible_positions += accessible_positions_this_loop
    


print(total_accessible_positions)
```
