### Question 5
https://adventofcode.com/2025/day/5

```python
range_inputs = []
ingredients = []
on_range_input_stream = True
with open('question_5_input.txt', 'r') as f:
    all_joltages = []
    for line in f:
        if line.strip() == '':
            on_range_input_stream = False
            continue
        if on_range_input_stream:
            range_inputs.append(tuple(int(i) for i in line.strip().split("-")))
        else:
            ingredients.append(int(line.strip()))
        
```

```python
num_fresh_ingredients = 0
for ingredient in ingredients:
    for range_input in range_inputs:
        if range_input[0] <= ingredient <= range_input[1]:
            num_fresh_ingredients += 1
            break
num_fresh_ingredients
```

```python
def count_distinct_numbers_in_ranges(ranges):
    """
    Count the number of distinct numbers across all ranges.
    
    Args:
        ranges: List of tuples (start, end) where both are inclusive
        
    Returns:
        int: Number of distinct numbers across all ranges
    """
    if not ranges:
        return 0
    
    # Sort ranges by start position
    sorted_ranges = sorted(ranges)
    
    # Merge overlapping or adjacent ranges
    merged = [sorted_ranges[0]]
    
    for current_start, current_end in sorted_ranges[1:]:
        last_start, last_end = merged[-1]
        
        # If current range overlaps or is adjacent to the last merged range
        if current_start <= last_end + 1:
            # Merge by extending the end if necessary
            merged[-1] = (last_start, max(last_end, current_end))
        else:
            # No overlap, add as new range
            merged.append((current_start, current_end))
    
    # Count distinct numbers in merged ranges
    total = sum(end - start + 1 for start, end in merged)
    
    return total
```

```python
count_distinct_numbers_in_ranges(range_inputs)
```
